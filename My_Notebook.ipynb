# Load the libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Load the CSV files

```python
gas = pd.read_csv("gas_prices.csv")

plt.figure(figsize=(8,5))
plt.plot(gas.Year, gas.USA, "b.-", label="USA",)
plt.plot(gas.Year, gas.Canada, "r.-", label="Canada")

# If a name has multiple words the (Dot) format won't work.
# Do this instead
plt.plot(gas["Year"],gas["South Korea"], "g.-", label="South Korea")
plt.plot(gas["Year"],gas["Australia"], "y.-")

# Another way to plot values
# for country in gas:
#     if country != "Year":
#         plt.plot(gas["Year"],gas[country], marker=".",label=country)
plt.title("Gas Prices Over Time (USD)", fontdict={"fontweight":"bold", "fontsize": 18})
plt.xlabel("Year")
plt.ylabel("Price per gallon")
plt.legend()

# Change the ticks
plt.xticks(gas.Year[::3].tolist() + [2011])

# Save graph
plt.savefig("gas_prices_over_time.png",dpi=300)

plt.show()
```

## Histogram Example

```python
# Load FIFA Data
fifa = pd.read_csv("fifa_data.csv")
fifa
```

### Histograms

```python
bins = [number for number in range(40,110,10)]
# How many players are in 40-50 range, 50-60 range, etc.

plt.hist(fifa.Overall, bins=bins, color="#6ceb5b")
plt.xticks(bins)
plt.xlabel("Skill level")
plt.ylabel("Number of Players")
plt.title("Distribution of Player SKills in FIFA 2018")
plt.show()
```

### Pie Charts

```python
left = fifa.loc[fifa["Preferred Foot"] == "Left"].count()[0]
right = fifa.loc[fifa["Preferred Foot"] == "Right"].count()[0]

labels = ["Left", "Right"]
colors = ["#3265c9", "#9734b3"]

plt.pie([left, right], labels=labels, colors=colors, autopct="%.2f %%")

plt.title("Foot Preference of FIFA Players")

plt.show()
```

```python
fifa.Weight = [int(x.strip("lbs")) if type(x)==str else x for x in fifa.Weight]

# Change style of plot ahead of time. There's a list you can find.
plt.style.use("ggplot")

fifa.Weight[0]

light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa.loc[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa[(fifa.Weight >= 200)].count()[0]

weights = [light, light_medium, medium, medium_heavy, heavy]
labels = ["Under 125", "125-150", "150-175", "175-200", "Over 200"]
explode = (.4,.2,0,0,.4)

plt.pie(weights, labels = labels, autopct="%.2f %%", pctdistance = 0.8, explode = explode)
# pctdistance is from 0-1
# Explode separates a set of data from the rest of the data (distance wise)

plt.title("Weight Distribution of FIFA Players (in lbs)")

plt.show()
```

### Box and Whiskers

```python
plt.style.use("default")

plt.figure(figsize=(5,8))

barcelona = fifa.loc[fifa.Club == "FC Barcelona"]["Overall"]
madrid = fifa.loc[fifa.Club == "Real Madrid"]["Overall"]
revs = fifa[fifa.Club == "New England Revolution"]["Overall"]

labels = ["FC Barcelona", "Real Madrid", "NE Revolution"]
boxes = plt.boxplot([barcelona, madrid, revs], labels=labels ,patch_artist=True, medianprops={"linewidth":2})

for box in boxes["boxes"]:
    # Set Edge Color
    box.set(color = "#4286F4", linewidth = 2)
    
    # Change Fill Color. Make sure to have patch_artist = True in the boxes 
    box.set(facecolor = "#e0e0e0")
    

plt.title("Professional Soccer Team Comparison")
plt.xlabel("Teams")
plt.ylabel("FIFA Overall Rating")

plt.show()
```
